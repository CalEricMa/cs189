```python
import numpy as np
import matplotlib.pyplot as plt

def p_norm(x, y, p):
    return (np.abs(x)**p + np.abs(y)**p)**(1/p)

x_vals = np.linspace(-5, 5)
y_vals = np.linspace(-5, 5)
x, y = np.meshgrid(x_vals, y_vals)

p_values = [0.5, 1, 2]
for p in p_values:

    z = p_norm(x, y, p)

    plt.figure(figsize=(5,5))
    contour = plt.contour(x, y, z)

    plt.title(p)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.axis("equal")
    plt.show()
```
